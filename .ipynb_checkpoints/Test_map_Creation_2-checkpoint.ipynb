```python
T = fun.getAllBarrisBCNPoligonBox()
map = fun.mapCreation(41.388790,2.158990)
cluster = fun.mapWithMarkerCluster(map, "Restaurants Barcelona")
porc = 0
maxP = len(T)
```

```python
map = fun.mapCreation(41.388790,2.158990)
x = 0
maxSee = 30
porc = 0
for barri in T: #all barcelona polygons
    if x < maxSee:
        fun.mapAddLine(map,fun.transformArrYXToXY(T[barri]), "#0000FF", 2.5, 1)
        n = fun.getAmenityInfoIntoPolygon(T[barri], "restaurant")
        data = fun.getInfoOfOSMSearch(n)
        for d in data:
            tmp = d["properties"]          
            name = ""
            if 'name' in tmp:
                name = tmp["name"]
            fun.mapAddMarker(map, d["geometry"][0], d["geometry"][1], 'glyphicon-glass', "green", name)
        porc +=1
        print (porc * 1.0 / min(maxP, maxSee) * 1.0)* 100
    x += 1
```
