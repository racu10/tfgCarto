```python
import pandas as pd
import numpy as np
import overpy
import overpass
import folium
from osmapi import OsmApi
import math
```

```python
def getDistance(long1, lat1, long2, lat2):
    r = 6371000 #radio terrestre medio, en metros 
    c = math.pi/180 #constante para transformar grados en radianes 
    return 2*r*math.asin(math.sqrt(math.sin(c*(lat2-lat1)/2)**2 + math.cos(c*lat1)*math.cos(c*lat2)*math.sin(c*(long2-long1)/2)**2))
```

```python
allCorrd = allData['features']
```

```python
map2 = folium.Map(location=[allData["features"][0]['geometry']['coordinates'][0][1],allData["features"][0]['geometry']['coordinates'][0][0]])
#markerCluster2 = folium.MarkerCluster("pubs").add_to(map2)


for geo in allCorrd:
    geometry = geo["geometry"]
    
    #print(geo, "\n")
    if geo["properties"]["name"].upper() == "Passeig de Gràcia".upper():
        if geometry["type"].upper() == "LINESTRING":
            for c in geometry["coordinates"]:
                y = c[0]
                x = c[1]
                #print(x, ' - ', y)
                txtHtml = "x: "+ str(x) + ' - y: ' + str(y)
                poppin = folium.Popup(html=folium.element.IFrame(html=txtHtml,width=200,height=50))
                folium.Marker([x,y],popup = poppin).add_to(map2)
    
```

```python
testDist = (41.38911086884985, 2.1676714718341827)
```

```python
map = folium.Map(location=[allData["features"][0]['geometry']['coordinates'][0][1],allData["features"][0]['geometry']['coordinates'][0][0]])
folium.Marker([testDist[0],testDist[1]]).add_to(map)
minD = math.inf
cx = 0.0
cy = 0.0
for geo in allCorrd:
    geometry = geo["geometry"]
    
    #print(geo, "\n")
    if geo["properties"]["name"].upper() == "Passeig de Gràcia".upper():
        if geometry["type"].upper() == "LINESTRING":
            for c in geometry["coordinates"]:
                y = c[0]
                x = c[1]
                d = getDistance(testDist[0], testDist[1], x, y)
                #print(d)
                if d < minD:  
                    #print(d)
                    cx = x
                    cy = y 
                    minD = d
                    

folium.Marker([cx,cy]).add_to(map)
    
```
